# Importing DataSet

```python
import numpy as np
from collections import OrderedDict
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## Reading CSV

```python
data = pd.read_csv(r'C:\Users\deepa\Desktop\data.csv')
```

```python
df = pd.DataFrame(data)
print(df.shape)
```

### Removing Unnesscary Columns

```python
df.drop(columns=['URL','ID','Description','Primary Genre','Icon URL','Developer','Subtitle'],inplace=True)
print(df.shape)
```

### Filling empty with null Values

```python
df['In-app Purchases'].fillna(0,inplace=True)
df
```

### Filling The Empty Values of App Purchases

```python
l = [sum(map(float,i.split(','))) if type(i) == str else 0 for i in df['In-app Purchases'] ]#Adding all the values if present in aIn app purchases column
len(l)
df['In-app Purchases'] = l
df
```

### Removing all the User Rating Count with less than 200

```python
df = df[df['User Rating Count'] > 199]
df
```

```python
df.dropna(axis=0,subset=['Average User Rating'],inplace=True)
m = [x.replace(" ","") for x in df['Genres'] ]#Removing all the spaces from the Genre strings
df['Genres'] = m
df
```

### Update Date

```python
x=[]#Checking if the date is > 6 Months 
y=[]
def dateset(i):
    x = i.split('/')
    if ((int(x[1]) <= 10 and int(x[2]) == 2019) | int(x[2]) <= 2018) :
        return False
    else:
        return True
c = -1
for i in df['Current Version Release Date']:
    c+=1
    if dateset(i):
        y.append(df.index[c])
df.drop(y,inplace=True)
df
```

### Removing games and Entertainment from the data Frame

```python
p = [w.split(',') for w in df['Genres']]

#le = [x.remove(' Games') for x in p if ' Games' in x]
le1 = [x.remove('Games') for x in p if 'Games' in x]
#le2 = [x.remove(' Entertainment') for x in p if ' Entertainment' in x]
le3 = [x.remove('Entertainment') for x in p if 'Entertainment' in x]
print(len(p))
p
```

### Replacing Genres with Genrality

```python

real = [','.join(i) for i in p]#Reverting back to the native

real = [word.replace('Board','Puzzle') if 'Board' in word else word for word in real ]


real = [word.replace('RolePlaying','Adventure') if 'RolePlaying' in word else word for word in real ]
real = [word.replace(' Role',' Adventure') if ' Role' in word else word for word in real ]
#real = [word.replace(' Education',' Family') if ' Education' in word else word for word in real ]
real = [word.replace('Education','Family') if 'Education' in word else word for word in real ]
print(len(real))
real
```

### Replacing the Data Frame (Genre Column)

```python
df['Genres'] = real
df
```

### Different Types Of Genres

```python
d = {}
d1 = {}
l1 = []
for st in df['Genres']:
    s = st.split(',')
    for i in s:
        d = {i:0}
        l1.append(i)
        d1.update(d)
d2 = OrderedDict(sorted(d1.items()))

d2
```

### List of Tuples

```python
df['Tuples'] = list(zip(df.Genres, df['Average User Rating']))#Adding a Column for convienvce
df
```

### Finding Strategy Combination games

```python

for i,k in df['Tuples']:
    for j in i.split(','):
        if 'Strategy' in i:
            d2[j] += k - 3.9#Adding My own weight to each Genre depending upon the Avg Rating
d2

del d2['Strategy']
```
